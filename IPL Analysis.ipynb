```python
#Filter only last 4 IPL season
# --- Step 1: Extract the first year from "2020/21" format ---
import re
def extract_first_year(value):
    if pd.isna(value):
        return None
    s = str(value)
    # look for first 4-digit year (e.g. "2020" in "2020/21")
    match = re.search(r'(20\d{2})', s)
    if match:
        return int(match.group(1))
    return None

data_matches['season_year'] = data_matches['season'].apply(extract_first_year)
```

```python
recent_matches = data_matches[data_matches['season_year'].between(2020, 2024)]
print(f"Filtered recent matches: {recent_matches.shape[0]} rows")
print(recent_matches['season_year'].unique())
```

```python
recent_data = data_deliveries.merge(recent_matches[['id', 'season_year', 'city', 'winner']], left_on='match_id', right_on='id', how='left')
recent_data.tail(400)
```

```python
#batsmen analysis
# 1. Top batsmen per season

season_bat_stats = recent_data.groupby(['season_year','batter']).agg(
            total_runs = ('batsman_runs','sum'),
            ball_faced = ('ball','count'),
            matches_played = ('match_id',pd.Series.nunique)).reset_index()


season_bat_stats['strike_rate'] = (season_bat_stats['total_runs']/season_bat_stats['ball_faced']) * 100

season_bat_stats = season_bat_stats[season_bat_stats['matches_played'] >=3]

top_batsman_per_season = (
                           season_bat_stats
                           .sort_values(['season_year','total_runs'],ascending = [True,False])
                           .groupby('season_year')
                           .head(10))

top_batsman_per_season
```

```python
#Consistent players
consistency = top_batsman_per_season.groupby('batter')['season_year'].nunique().reset_index()
consistent_player = consistency[consistency['season_year']>1].sort_values(['season_year'],ascending= False)

print('Players consistently in top 10 across seasons:\n',consistent_player)
```

```python
#Top bowlers 
# Group by season and bowler
season_bowl_stats = recent_data.groupby(['season_year', 'bowler']).agg(
    runs_conceded = ('total_runs', 'sum'),
    balls_bowled = ('ball', 'count'),
    wickets = ('player_dismissed', lambda x: x.notna().sum()),
    matches_played = ('match_id', pd.Series.nunique)
).reset_index()

# Filter: bowlers who played at least 3 matches
season_bowl_stats = season_bowl_stats[season_bowl_stats['matches_played'] >= 3]

# Calculate overs bowled (convert balls to overs)
season_bowl_stats['overs'] = season_bowl_stats['balls_bowled'] // 6 + (season_bowl_stats['balls_bowled'] % 6) / 6

# Economy rate = runs per over
season_bowl_stats['economy'] = season_bowl_stats['runs_conceded'] / season_bowl_stats['overs'].replace(0, np.nan)

# Bowling strike rate = balls per wicket
season_bowl_stats['bowling_strike_rate'] = season_bowl_stats['balls_bowled'] / season_bowl_stats['wickets'].replace(0, np.nan)

# Sort to get top bowlers by wickets each season
top_bowlers_per_season = (
    season_bowl_stats
    .sort_values(['season_year', 'wickets', 'economy'], ascending=[False, False, True])
    .groupby('season_year')
    .head(10)
)

top_bowlers_per_season.head(15)
```

```python
#Top All Rounder
# Rename columns to have consistent key 'player'
season_bat_stats = season_bat_stats.rename(columns={'batter': 'player'})
season_bowl_stats = season_bowl_stats.rename(columns={'bowler': 'player'})

# Merge both batting and bowling
season_allrounder_stats = pd.merge(
    season_bat_stats,
    season_bowl_stats,
    on=['season_year', 'player'],
    how='inner'  # only players who batted AND bowled
)

season_allrounder_stats.head()
```

```python
# Batting impact: prioritize runs and strike rate
season_allrounder_stats['batting_score'] = (
    0.6 * season_allrounder_stats['total_runs'] +
    0.4 * season_allrounder_stats['strike_rate']
)

# Bowling impact: prioritize wickets and economy (lower economy = better)
season_allrounder_stats['bowling_score'] = (
    0.7 * season_allrounder_stats['wickets'] +
    0.3 * (1 / season_allrounder_stats['economy'].replace(0, np.nan))
)

# Combined all-rounder performance score
season_allrounder_stats['allrounder_score'] = (
    0.5 * season_allrounder_stats['batting_score'] +
    0.5 * season_allrounder_stats['bowling_score']
)
season_allrounder_stats
```

```python
top_allrounders_per_season = (
    season_allrounder_stats
    .sort_values(['season_year', 'allrounder_score'], ascending=[False, False])
    .groupby('season_year')
    .head(10)
)

top_allrounders_per_season[['season_year', 'player', 'total_runs', 'wickets', 'strike_rate', 'economy', 'allrounder_score']].head(15)
```

```python
#Combine all top scorer data in one dataframe

# rename to make one column 'player'
bats = top_batsman_per_season.rename(columns={'batter': 'player'})
bowls = top_bowlers_per_season.rename(columns={'bowler': 'player'})
allr = top_allrounders_per_season.rename(columns={'player': 'player'})

# add a column for role
bats['role'] = 'Batsman'
bowls['role'] = 'Bowler'
allr['role'] = 'All-Rounder'

# combine all
players = pd.concat([bats, bowls, allr], ignore_index=True)

# keep only needed columns
players = players[['season_year', 'player', 'role', 'total_runs', 'strike_rate', 'wickets', 'economy']]
players.fillna(0, inplace=True)
players.head()
```

```python
def normalize(s):
    return (s - s.min()) / (s.max() - s.min())

players['norm_runs'] = normalize(players['total_runs'])
players['norm_sr'] = normalize(players['strike_rate'])
players['norm_wickets'] = normalize(players['wickets'])
players['norm_econ'] = normalize(1 / (players['economy'] + 0.1))  # lower economy = better
```

```python
def performance_score(row):
    if row['role'] == 'Batsman':
        return 0.6 * row['norm_runs'] + 0.4 * row['norm_sr']
    elif row['role'] == 'Bowler':
        return 0.7 * row['norm_wickets'] + 0.3 * row['norm_econ']
    else:  # All-Rounder
        bat_score = 0.5 * row['norm_runs'] + 0.3 * row['norm_sr']
        bowl_score = 0.5 * row['norm_wickets'] + 0.2 * row['norm_econ']
        return 0.5 * bat_score + 0.5 * bowl_score

players['performance_score'] = players.apply(performance_score, axis=1)
players.head()
```

```python
consistency = players.groupby('player')['season_year'].nunique().reset_index()
consistency.columns = ['player', 'season_count']
players = players.merge(consistency, on='player', how='left')
players['consistency_score'] = players['season_count'] / players['season_count'].max()
```

```python
players['final_score'] = (0.85 * players['performance_score']) + (0.15 * players['consistency_score'])
players = players.sort_values('final_score', ascending=False)
players.head()
```

```python
best_batsmen = players[players['role'] == 'Batsman'].sort_values('final_score', ascending=False).head(4)
best_allr = players[players['role'] == 'All-Rounder'].sort_values('final_score', ascending=False).head(2)
best_bowlers = players[players['role'] == 'Bowler'].sort_values('final_score', ascending=False).head(4)

# pick 1 extra best from remaining players (any role)
remaining = players[~players['player'].isin(pd.concat([best_batsmen, best_allr, best_bowlers])['player'])]
extra = remaining.sort_values('final_score', ascending=False).head(1)

# final team
best_team = pd.concat([best_batsmen, best_allr, best_bowlers, extra])
best_team[['player', 'role', 'season_year', 'final_score']].reset_index(drop=True)
```

```python
#Final Team Chart
sns.barplot(data=best_team, x='final_score', y='player', hue='role')
plt.title("Best Franchise Team (2021–2024 Analysis)")
plt.xlabel("Final Score (Performance + Consistency)")
plt.ylabel("Player")
plt.show()
```
